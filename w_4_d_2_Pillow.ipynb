```
from __future__ import print_function
import os, sys
from PIL import Image

for i in sys.argv[1:]:
  f,e=os.path.splitext(i)
  o=f+'.jpg'
  if i != o:
    try:
      Image.open(i).save(o)
    except IOError:
      print("cannot convert",i)
```
